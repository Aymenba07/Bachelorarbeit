```python
def createSubDataframe(df,**config):
    columns = list(config2["columns"].keys())
    if 'first' in columns:
        columns.remove('first') 
    if 'last' in columns:
        columns.remove('last') 
    df_new = df[columns]
    for key,value in config2["columns"].items():
        print(key,value)
        if key =="first":
            df_new =  df_new.head(value).reset_index(drop=config["reset_index"])
        elif key == "last":
            df_new =  df_new.head(value).reset_index(drop=config["reset_index"])
        elif value is not None:
            if value[0] == "max":
                df_new =  df_new.nlargest(value[1], key)
            elif value[0] == "min":
                df_new = df_new.nsmallest(value[1], key)
            elif value[0] == "first":
                df_new = df_new.head(value[1])
            elif value[0] == "last":
                df_new = df_new.tail(value[1])
            elif value[0] == "equal":
                if isinstance(value[1],list):
                    df_new = df_new[df_new[key].isin(value[1])]
                else:
                    df_new = df_new.loc[df_new[key] == value[1]]
            elif value[0] == "interval":
                df_new = df_new.loc[(df_new[key] >= value[1][0]) & (df_new[key] <= value[1][1])]
        print(df_new)
    return df_new.reset_index(drop=config2["reset_index"])
```
